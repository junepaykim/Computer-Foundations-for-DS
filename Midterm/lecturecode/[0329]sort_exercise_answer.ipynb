## **Exercise** : Basic Sorting Algorithm - selection sort and insertion sort 
##### Reference : COCI 2014/2015 Task CESTA 4th round, December 20th, 2014 



#### 어느 날, 현우는 우연히 양수 N이 적힌 카드들을 보았다. 현우는 30이란 수를 좋아하기 때문에, 그는 카드에 포함된 숫자들을 섞어 30의 배수가 되는 가장 큰 수를 만들고 싶어한다. 현우가 만들고 싶어하는 수를 계산하는 코드를 작성하라.
#### One morning, completely by chance, Hyunwoo found cards with positive integers N. Since Hyunwoo adores the number 30, he wants to know the maximum multiple of the number 30 that can be obtained by shuffling the cards with the digits of the number he found. Help Hyunwoo and write a code that calculates that number (if it exists).


##### [Conditions]
##### 입력 (INPUT) : N을 입력받는다. N는 최대 100,000개의 숫자로 구성되어 있으며, 0으로 시작하지 않는다.(The first and only line of input contains the integer N, consisting of at most 100,000 digits.)
##### 출력 (OUTPUT) : 만들고 싶어하는 수가 존재한다면 그 수를 출력하라. 그 수가 존재하지 않는다면, -1을 출력하라. (The first and only line of output must contain the required number from the task, if it exists. If it doesn’t exist, output -1.)

**Examples**

| input example  |  output example |
| :--: | :--: |
|    30  |   30  |
|   102 | 210  |
|   2931   |   -1   |
|   80875542   |    88755420  |

### insertion sort

```python
# divisibility rules for 30?
def insertion_sort(N: int) :
    L = list(str(N))
    L = list(map(int, L))
    for i in range(1, len(L)):
        for j in range(i, 0, -1):
            if L[j-1] < L[j]:
                L[j-1], L[j] = L[j], L[j-1]
            else:
                break

    if L[-1] == 0 and sum(L) % 3 == 0:
        f_list = list(map(str, L))
        return int("".join(f_list))
    else:
        return -1
```

```python
insertion_sort(102)
```

```python
insertion_sort(2931)
```

### selection sort

```python
def selection_sort(N:int):
    L = list(str(N))
    L = list(map(int, L))
    for i in range(len(L)):
        largest = i
        for j in range(i+1, len(L)):
            if L[j] > L[largest]:
                largest = j
        L[i], L[largest] = L[largest], L[i]
    
    if L[-1] == 0 and sum(L) % 3 == 0:
        f_list = list(map(str, L))
        return int("".join(f_list))
    else:
        return -1
```

```python
selection_sort(102)
```

```python
selection_sort(2931)
```
