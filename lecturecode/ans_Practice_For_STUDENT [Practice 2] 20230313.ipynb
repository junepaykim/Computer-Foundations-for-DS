# Q1. Beef Roasting

### Referring to a beef temperature chart below, write a code that takes internal_temperature as an input and outputs the roasted degree.

```python
def roasting (internal_temperature: int) -> str:
    if 45<=internal_temperature<=47 :
        roasted = "Rare"
    elif 50<=internal_temperature<=52 :
        roasted = "Medium rare"
    elif 55<=internal_temperature<=60 :
        roasted = "Medium"
    elif 64<=internal_temperature<=70 :
        roasted = "Well done"
    else:
        roasted = "Undefined"
    return roasted

print(roasting(50))
print(roasting(68))
print(roasting(100))
```

# Q2. Name Separator

### Taking a name as input, write a function that separates last name and first name.

```python
def name_separator(name):
    
    #split first name and last name 
    separated = name.split()
    first_name = separated[0]
    last_name = separated[-1]
        
    print(f"first name is {first_name}")
    print(f"last name is {last_name}")

my_name = input("Write your name:(eg.Sun-sin Yi) ")
name_separator(my_name)
```

# Q3. Age Calculator

### Taking the first seven digits of the residential number as input values, write a function that outputs the age and gender.

**Guidelines** 
1) Distinguish gender (hint: 7th digit: 1 or 3- male, 2 or 4-female) \
2) Calculate the age based on today's date (hint:compare birthday and today's date)

```python
from datetime import date
today = date.today()
today = today.strftime("%Y%m%d")
    

def q3(personal_number):
    
    age=0
    gender=''
    
    #split numbers for its purpose (age, gender)
    num4age = personal_number[:6] #str
    gender_code=personal_number[-1] #str
    
    #step1) gender 
    if gender_code in('1','3'):
        gender='male'
    else:
        gender='female'  
        
    #step2)
    #changing format to calculate age
    if gender_code in('3','4'):
        birth ='20'+ num4age
    else :
        birth ='19'+ num4age
        
    
    ##cases that birthday has not passed
    if int(birth[4:]) > int(today[4:]):
        age=int(today[:4])-int(birth[:4])
        
    ##cases that birthday has passed(+1)
    else:
        age=int(today[:4])-int(birth[:4])+1
            
    print(f"Gender: {gender}")
    print(f"Age: {age}")

my_number = input("Write your personal number:(eg.220908-2)  ")
q3(my_number)
```

# Q4. Caesar Cipher

### A method of encryption that changes each alphabet of a sentence to another alphabet by pushing it by a certain distance is called Caesar Cipher.

### For example, "A" becomes "B" when pushed by 1, and "D" when pushed by 3. If you push "z" by 1, it becomes "a". Write a code that takes the string s and the distance n and creates a ciphertext pushed by n.

**restriction** 
1) only single character is considered as input \
2) The blank space is blank no matter how much you push it \
3) s is lowercase letter or uppercase letter or space only \
4) n is a natural number (1<=n<=25)\
-reference: programmers

##### solution

```python
def caesar_cipher(s:str, n:int)->str:
    ans=''
    if s.isupper():
        ans = chr((ord(s) - ord('A') + n) % 26 + ord('A'))        
    elif s.islower():
        ans= chr((ord(s) - ord('a') + n) % 26 + ord('a'))
    else:
        pass
    
    return ''.join(ans)


print(caesar_cipher('',2))
print(caesar_cipher('A',2))
print(caesar_cipher('z',3))
```
