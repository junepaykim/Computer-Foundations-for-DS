### Q1. Bubble Sort

[KOR] 버블 정렬(bubble sort)은 인접한 두 원소의 순서가 바뀌어 있을 때 (앞의 원소가 뒤의 원소보다 클 때) 서로 swap하면서 정렬하는 알고리즘이다. 정수로 이루어진 리스트(list)를 변수(argument)로 받아 버블 정렬할 때, swap이 일어나는 횟수를 리턴하는 함수를 작성하시오.

[ENG] Bubble Sort is the simplest sorting algorithm that works by repeatedly swapping the adjacent elements if they are in wrong order. Return the number of swaps when bubble sorting a list of integers.

- ex1 - BubbleSort([5,1,4,2,8])
> 4
- ex2 - BubbleSort([-1, -1, -1, -1, -1]) 
> 0
- ex3 - BubbleSort([6, 5, 4, 3, 2, 1]) 
> 15
- ex4 - BubbleSort([1, 5, 2, 3, 6, 6, 1, 2, 3, 4, 21, 1, 11, -1]) 
> 40

#### Solution

```python
def BubbleSort(num_list: list) -> int:

    ### Write code here ###
    ans = 0
    for i in range(len(num_list) - 1) :
        for j in range(len(num_list) -i - 1) :
            if num_list[j] > num_list[j+1] :
                num_list[j], num_list[j+1] = num_list[j+1], num_list[j]
                ans += 1
    return ans
    
    ### End of your code ###  
```
