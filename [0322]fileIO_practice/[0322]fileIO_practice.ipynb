# Practice 1 - Cafe menu


### 1) Output: list of lists


```python
from typing import TextIO, List


def menu(filename: TextIO) -> List[List]:
    ##### Write your Code Here #####
    file = open(filename, "r")
    lines = [line.rstrip() for line in file.readlines()]
    file.close()
    return lines
    ##### End of your code #####
```

```python
menu("menu.txt")
```
